```python
# Check if there are difference between "year" column and the year in the "released" year

movies[~movies.apply(lambda x: str(x.year) in str(x.released), axis=1)]
```

```python
# In 708 rows, the value in "year" column and the year in the "released_date" column are different. 
# To avoid confusion, we will drop the "year" column 
# then take the year out from the "released_date" column and create a new column indicating the released year.

released_date = []
released_country = []
released_year = []

for x in movies['released']:
    date = x.split("(")[0]
    country = x.split("(")[1].replace(')','')
    released_date.append(date)
    released_country.append(country)
    
movies['released_country'] = released_country
movies['released_date'] = released_date
movies['released_date'] = pd.to_datetime(movies['released_date'])

movies['released_year'] = movies['released_date'].dt.year
movies['released_month'] = movies['released_date'].dt.month    
```
